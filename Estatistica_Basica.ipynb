```python
def imc(peso, altura):
  if peso <= 0:
    print("Por favor verifique o peso informado")
  if altura <= 0.2:
    print("Por favor verifique a altura informada")

  if peso <= 0 or altura <= 0.2:
    return None, "Dados invÃ¡lidos"

  indice = round(peso / (altura ** 2 ), 2)

  if indice <= 18.5:
    classificacao = "Abaixo do peso"
  elif indice <= 24.99:
    classificacao = "Peso normal"
  elif indice <= 29.99:
    classificacao = "Sobrepeso"
  else:
    classificacao = "Obesidade"

  return indice, classificacao
```

```python
indice, classificacao = imc(94,1.74)
print(indice, classificacao)
```
